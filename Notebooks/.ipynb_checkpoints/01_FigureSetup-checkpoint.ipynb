### One patient in the Bi & He cohort, P915 was misannotated as Stage IV. The biopsy was taken at Stage III disease. We first correct this in the full dataset Seurat object. <a name="1"></a>

```r
seur_full = readRDS("../../BraunBi_2022_RCC/BBK_Manuscript_Annotation/seur_goodcells_BBK_Manuscript.RDS")
table(seur_full$Sample)
table(seur_full$Stage)
```

```r
seur_full_meta = seur_full@meta.data
seur_full_meta = mutate(seur_full_meta, Stage_Corrected = ifelse(Sample %in% c("Bi_0600915_T1_TrueTumor"), "III", Stage))
seur_full_meta = mutate(seur_full_meta, Stage_Corrected_EarlyLate = ifelse(Stage_Corrected %in% c("I","II"), "Early",ifelse(Stage_Corrected %in% c("III","IV"),"Late","NotAvail")))

table(seur_full_meta$Stage)
table(seur_full_meta$Stage_Corrected)
table(seur_full_meta$Stage_Corrected_EarlyLate)
```

```r
table(subset(seur_full@meta.data, Stage_Corrected == "III")$Sample)
table(subset(seur_full@meta.data, Stage_Corrected == "IV")$Sample)
```

```r
table(subset(seur_full@meta.data, Stage_Corrected_EarlyLate == "Early")$Sample)
table(subset(seur_full@meta.data, Stage_Corrected_EarlyLate == "Late")$Sample)
```

```r
io_tx_meta = seur_full@meta.data
io_tx_meta = mutate(io_tx_meta, IO_Treated = ifelse(IO %in% c("NE","PD","PR","SD","Yes"),"IO Exposed","IO Naive"))
corner(io_tx_meta)
table(io_tx_meta$IO_Treated)

seur_full$IO_Treated = io_tx_meta$IO_Treated
```

### Define a set of putative ISGs. We define this is genes in Hallmark IFNa response, Hallmark IFNG response, Minn ISG.RS, Minn IFNG.GS, and a curated list of known immune checkpoints, most with proven IFN-responsiveness. <a name="2"></a>

```r
input <- scan("../Signatures/h.all.v2023.2.Hs.symbols.gmt", what="", sep="\n")
# Separate elements by one or more whitepace
hallmark <- strsplit(input, "[[:space:]]+")
# Extract the first vector element and set it as the list element name
names(hallmark) <- sapply(hallmark, `[[`, 1)
#names(y) <- sapply(y, function(x) x[[1]]) # same as above
# Remove the first vector element from each list element
hallmark <- lapply(hallmark, `[`, -1)
hallmark <- lapply(hallmark, `[`, -1)

hallmark_ifna = hallmark$HALLMARK_INTERFERON_ALPHA_RESPONSE
hallmark_ifng = hallmark$HALLMARK_INTERFERON_GAMMA_RESPONSE

length(hallmark_ifna)
length(hallmark_ifng)
```

```r
input <- scan("../Signatures/Benci_Minn_ISG_Signatures.txt", what="", sep="\n")
# Separate elements by one or more whitepace
benci <- strsplit(input, "[[:space:]]+")
# Extract the first vector element and set it as the list element name
names(benci) <- sapply(benci, `[[`, 1)
#names(y) <- sapply(y, function(x) x[[1]]) # same as above
# Remove the first vector element from each list element
benci <- lapply(benci, `[`, -1)
benci <- lapply(benci, `[`, -1)
benci

ISG.RS = benci$Minn_Benci_ISG.RS
IFNG.GS = benci$Minn_Benci_IFNG.GS

length(ISG.RS)
length(IFNG.GS)
```

```r
saveRDS(benci, file = "../Signatures/Benci_Minn_ISG_Signatures.RDS")
```

### Define a set of variable genes from the full dataset to use in mixed models for discovering IFN-responsive genes and also differential expression analysis to identify cell type specific genes.<a name="3"></a>

```r
seur_full = readRDS("../Data_Objects/seur_full_BBK_Manuscript.RDS")
```

```r
Idents(seur_full) = "BBK_Broad_Celltype"
table(Idents(seur_full))
```

```r
seur_full = subset(seur_full, idents = c("Contam","Omitted"), invert = T)
```

### First, we get the top 2000 variable genes using all cells

```r
seur_full = FindVariableFeatures(seur_full, selection.method = "vst",nfeatures = 2000)
all_var2000 = VariableFeatures(seur_full)
all_var2000
```

### Next, we get the top 2000 variable genes within each broad cell type

```r
vars = as.character(unique(Idents(seur_full)))
vars

var2000_list = list()
for (i in 1:length(vars)){
    seur_sub = subset(seur_full, idents = vars[i])
    seur_sub = FindVariableFeatures(seur_sub, selection.method = "vst",nfeatures = 2000)
    var2000_list[[i]] = VariableFeatures(seur_sub)
}

names(var2000_list) = vars
```

```r
str(var2000_list)
```

```r
var2000_unlist = unique(unlist(var2000_list))
length(var2000_unlist)
corner(var2000_unlist)
```

```r
final_var2000 = unique(c(all_var2000,var2000_unlist))
length(final_var2000)
```

```r
saveRDS(final_var2000, file = "../Signatures/manuscript_all_broadcelltype_var2000_seur_full.RDS")
```

### Rscripts/Manuscript_MixedEffects_Patient_Stage_Corrected_EarlyLate/MixedEffects_Patient_Stage_Corrected_EarlyLate_Vargenes_1_3.R

```r
library(Seurat)
library(useful)
library(lmerTest)
library(dplyr)

setwd("/broad/VanAllenLab_CA/xchip/cga_home/kevinbi/IFN_RCC_Manuscript_2024")

Vargenes = readRDS("Signatures/manuscript_all_broadcelltype_var2000_seur_full_notx_allstages.RDS")
length(Vargenes)
length(unique(Vargenes))

goi = readRDS("Signatures/manuscript_goi_hallmark_benciminn_curated.RDS")
length(goi)
length(unique(goi))

Vargenes = unique(c(goi, Vargenes))

seur_full_notx_allstages = readRDS("Data_Objects/seur_full_BBK_notx_allstages_No_Omitted_Contam_Manuscript.RDS")
Idents(seur_full_notx_allstages) = "BBK_Broad_Celltype_DE"

cytosig_outputs_seur_full_notx_allstages = readRDS("Data_Objects/cytosig_outputs_seur_full_notx_allstages.RDS")
celltypes = names(cytosig_outputs_seur_full_notx_allstages)

for (i in 1:3) {

test_cyto_pval = read.table(cytosig_outputs_seur_full_notx_allstages[[celltypes[i]]]["Pvalue"], header = T, row.names = 1, check.names = F, sep = '\t')
test_cyto_pval = data.frame(t(test_cyto_pval))
colnames(test_cyto_pval) = paste(colnames(test_cyto_pval),"_CytoSig_Pval",sep ="")
corner(test_cyto_pval)

test_cyto_zscore = read.table(cytosig_outputs_seur_full_notx_allstages[[celltypes[i]]]["Zscore"], header = T, row.names = 1, check.names = F, sep = '\t')
test_cyto_zscore = data.frame(t(test_cyto_zscore))
colnames(test_cyto_zscore) = paste(colnames(test_cyto_zscore),"_CytoSig_Zscore",sep ="")
corner(test_cyto_zscore)

seur_test = sub = subset(seur_full_notx_allstages, idents = celltypes[i])
seur_test@meta.data = cbind(seur_test@meta.data, test_cyto_zscore, test_cyto_pval)
dim(seur_test)
    
vec = as.vector(table(seur_test$Patient))
names(vec) = names(table(seur_test$Patient))
seur_test = subset(seur_test, subset = Patient %in% names(which(vec > 1)))
table(seur_test$Patient)
dim(seur_test)

Vargenes = intersect(Vargenes, row.names(seur_test@assays$RNA@data))
seur_test_Vargenes = as.matrix(seur_test@assays$RNA@data[Vargenes,])
seur_test_Vargenes = data.frame(t(seur_test_Vargenes),check.names = F)
seur_test_Vargenes = seur_test_Vargenes[,names(which(colSums(seur_test_Vargenes) != 0))]

corner(seur_test_Vargenes)
dim(seur_test_Vargenes)

new_Vargenes = colnames(seur_test_Vargenes)
length(new_Vargenes)

seur_test_meta = seur_test@meta.data[,c("Biopsy","Stage_Corrected_EarlyLate","Sample","SampleName","Patient","nFeature_RNA","nCount_RNA","IFN1_CytoSig_Zscore","IFNG_CytoSig_Zscore")]
head(seur_test_meta)

seur_test_df = cbind(seur_test_meta,seur_test_Vargenes)
seur_test_df$Patient = factor(seur_test_df$Patient)
seur_test_df$Sample = factor(seur_test_df$Sample)
seur_test_df$Stage_Corrected_EarlyLate = factor(seur_test_df$Stage_Corrected_EarlyLate)

res_beta = list()
res_pval = list()

for (k in 1:length(new_Vargenes)){
    sub = seur_test_df
    sub$test = as.numeric(seur_test_df[,new_Vargenes[k]])
    
    lm_res = lmer(scale(test) ~ scale(IFN1_CytoSig_Zscore) + scale(IFNG_CytoSig_Zscore) + scale(nFeature_RNA) + factor(Stage_Corrected_EarlyLate) + (1 | Patient), data = sub)
    lm_res_summ = summary(lm_res)
 
    res_beta[[k]] = lm_res_summ$coefficients[,"Estimate"]
    res_pval[[k]] = lm_res_summ$coefficients[,"Pr(>|t|)"]
}

names(res_beta) = new_Vargenes
names(res_pval) = new_Vargenes

res_beta_df = data.frame(t(as.data.frame(res_beta)), check.names = F)
res_beta_df = res_beta_df[,c("scale(IFN1_CytoSig_Zscore)","scale(IFNG_CytoSig_Zscore)")]
colnames(res_beta_df) = c("IFN1_resp_beta","IFNG_resp_beta")
row.names(res_beta_df) = new_Vargenes
head(res_beta_df)

res_pval_df = data.frame(t(as.data.frame(res_pval)), check.names = F)
res_pval_df = res_pval_df[,c("scale(IFN1_CytoSig_Zscore)","scale(IFNG_CytoSig_Zscore)")]
colnames(res_pval_df) = c("IFN1_resp_pval","IFNG_resp_pval")
row.names(res_pval_df) = new_Vargenes
head(res_pval_df)

res_plot = cbind(res_beta_df,res_pval_df)
res_plot$gene = row.names(res_plot)

res_plot$IFN1_resp_p_adj_BH = p.adjust(res_plot$IFN1_resp_pval,method = "BH")
res_plot$IFNG_resp_p_adj_BH = p.adjust(res_plot$IFNG_resp_pval,method = "BH")

res_plot$IFN1_resp_p_adj_Bonf = p.adjust(res_plot$IFN1_resp_pval,method = "bonferroni")
res_plot$IFNG_resp_p_adj_Bonf = p.adjust(res_plot$IFNG_resp_pval,method = "bonferroni")


head(res_plot)

saveRDS(res_plot, file = paste("/broad/VanAllenLab_CA/xchip/cga_home/kevinbi/IFN_RCC_Manuscript_2024/Outputs/Manuscript_MixedEffects_Output_Stage_Corrected_EarlyLate/",celltypes[i],"_CytoSig_MixedEffects_Patient_Stage_Corrected_EarlyLate_Vargenes.RDS", sep = "")) 

}
```
